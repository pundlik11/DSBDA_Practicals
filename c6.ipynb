```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data_airquality.csv', encoding='latin1')
```

```python
df = pd.read_csv('data_airquality.csv', encoding='latin1', low_memory=False)
```

```python
 df = df.apply(pd.to_numeric, errors='coerce')
```

```python
df.dtypes
```

```python
plt.figure(figsize=(8,6))
cols = ['so2', 'no2', 'rspm', 'spm']
corr = df[cols].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correaltion heatmap")
plt.show()
```

```python
sns.barplot(x='spm', y='rspm', data=df)
plt.show()
```

```python
#Histogram
df[["so2","no2","pm2_5"]].hist(bins=20, figsize=(12,6), layout=(1,3), color='skyblue')
plt.suptitle("histogram")
plt.tight_layout()
plt.show()
```

```python
for column in ["so2","no2","pm2_5"]:
    sns.kdeplot(data=df[column], fill=True)
    plt.title(f"KDE plot for {column}")
    plt.xlabel(column)
    plt.ylabel("Density")
plt.show()
```

```python
sns.scatterplot(x='spm', y='rspm', data=df)
plt.xlabel("spm")
plt.ylabel("rspm")
plt.title("Relationship between spm and rspm")
plt.show()
```

```python
sns.boxplot(x='so2', y='no2', data=df)
plt.title("Boxplot of so2 by presence")
plt.show()
```

```python
sns.countplot(x='pm2_5', data=df)
plt.title("Countplot of pm2_5")
plt.xlabel("pm2_5")
plt.ylabel("Count")
plt.show()
```

```python
sns.jointplot(data=df, x='spm', y='rspm', kind='scatter', hue='so2')
plt.suptitle('jointplot of spm vs rspm',y=1.02)
plt.show()
```
