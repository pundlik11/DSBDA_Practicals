```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
heart = pd.read_csv('dataset_Heart.csv')
```

```python
print("columns in the dataset:")
print(heart.columns)
```

```python
heart['binary_target'] = heart['target'].apply(lambda x: 0 if x ==0 else 1)
```

```python
sns.set(style="whitegrid")
```

```python
#Histogram
heart[["age","chol","thalach"]].hist(bins=20, figsize=(12,6), layout=
(1,3), color='skyblue')
plt.suptitle("histogram")
plt.tight_layout()
plt.show()
```

```python
#KDE plot
for column in ["age", "chol", "thalach"]:
 sns.kdeplot(data=heart[column], fill=True)
 plt.title(f"KDE plot for {column}")
 plt.xlabel(column)
 plt.ylabel("Density")
 plt.show()
```

```python
#Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heart.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation heatmap of heart dataset")
plt.show()
```

```python
sns.scatterplot(x='chol', y='age', hue="target", data=heart)
plt.xlabel("Chol")
plt.ylabel("age")
plt.title("Relationship between chol and age")
plt.show()
```

```python
sns.boxplot(x='target', y='chol', data=heart)
plt.title("Boxplot of cholesterol by heart disease presence")
plt.show()
```

```python
sns.pairplot(heart, hue='target')
plt.suptitle("Pairplot of Heart Dataset", y=1.02)
plt.show()
```

```python
sns.countplot(x='binary_target', data=heart)
plt.title("Countplot of Target (Heart Disease Present or Not)")
plt.xlabel("Target(1 = Disease, 0 = No Disease)")
plt.ylabel("Count")
plt.show()
```

```python
sns.jointplot(data=heart, x='age', y='chol', kind='scatter', hue='target')
plt.suptitle('jointplot of age vs chol',y=1.02)
plt.show()
```

```python
sns.violinplot(x='sexe', y='chol', data=heart)
```
